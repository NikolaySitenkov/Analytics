## Загрузка данных

```python
connection = {'host': 'https://karpov_courses_clickhouse_db',
              'database':'simulator_20221220',
              'user':'user_name', 
              'password':'password'
             }

query = '''
SELECT user_id,
       exp_group,
       sum(action = 'view') AS views,
       sum(action = 'like') AS likes,
       likes / views as ctr   
FROM {db}.feed_actions
WHERE toDate(time) BETWEEN '2022-11-25' AND '2022-12-01'
      AND exp_group IN (2, 3)
GROUP BY exp_group, user_id
'''

df_ctr = ph.read_clickhouse(query, connection=connection)
# Пришлось отсортировать по user_id, так как на стороне запроса user_id 
# перемешиваются каждый раз по разному и через np.random.seed не выходило
# получить тот же самый результат, так как когда отбирались 500 данных,
# строки постоянно имели разный порядок при запросах.
df_ctr = df_ctr.sort_values(by="user_id")
```

## A/A тест

```python
np.random.seed(10)

list_pvals = []
list_dec = []

gr_2 = df_ctr[df_ctr.exp_group == 2].ctr.values
gr_3 = df_ctr[df_ctr.exp_group == 3].ctr.values

for i in range(10000):
        
    gr_2_sample = np.random.choice(gr_2, size=500, replace=False)
    gr_3_sample = np.random.choice(gr_3, size=500, replace=False)

    pval = stats.ttest_ind(gr_2_sample, 
                           gr_3_sample, 
                           equal_var=False).pvalue
    
    list_pvals.append(pval)
    list_dec.append(int(pval <= 0.05))
    

res_pvals = np.array(list_pvals)
res_dec = np.array(list_dec)
```

```python
sns.set(
    font_scale=1,
    style="whitegrid",
    rc={"figure.figsize": (10, 7)}
)

ax = sns.histplot(data=pd.DataFrame(res_pvals, columns=["pvalue"]))
ax.set(title="Гистограмма со значениями p-value")
ax.set(xlabel="p-value", ylabel="Частота встречаемости")
plt.show()
```

```python
sns.set(
    font_scale=1,
    style="whitegrid",
    rc={"figure.figsize": (5, 5)}
)


data_dec = pd.DataFrame([("reject" if i else "accept") for i in res_dec], 
                         columns=["decision"])

ax = sns.countplot(data=data_dec, x="decision")
ax.set(title="Гистограмма по принятию решений")
ax.set(xlabel="decision", ylabel="Частота встречаемости")
plt.show()
```

```python
res_val = res_dec.sum() / res_dec.shape[0]
print(f"Процент p-values: {round(res_val * 100, 2)}")
```

**Вывод**
<br/>
<br/>
4.75% p-values оказался меньше либо равный 5%(0.05), соотвественно разницы между группами 2 и 3 нет, следовательно наша система сплитования работает корректно.
